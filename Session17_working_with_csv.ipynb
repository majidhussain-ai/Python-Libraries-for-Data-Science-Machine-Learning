### Perform parameters with different dataset to handle:

### 1.sep parameter.

```python
# open csv file with sep parameter if in our dataset have issue like this
df = pd.read_csv('session_27_dataset/movie_titles_metadata.tsv')
df.head()
```

```python
# Now I can solve this issue by using of sep parameter
df = pd.read_csv('session_27_dataset/movie_titles_metadata.tsv', sep = '\t')
df.head()
```

### 2. names parameter.

```python
# till have an issue that not clear our column name i can also solve this by using of names parameter and give names to seeing of dataset
df = pd.read_csv('session_27_dataset/movie_titles_metadata.tsv', sep = '\t', names = ['sno','name','release_year','rating','votes','genres'])
df.head()
```

```python
# Now if I want to give column name from dataset i can also  do this by using of col_name parameter
df = pd.read_csv('session_27_dataset/movie_titles_metadata.tsv', sep = '\t', names = ['sno','name','release_year','rating','votes','genres'])
df.head()
```

### 3. index_col paramter.

```python
# How to handle index from dataset. I can also choose an index from dataset like this
ipl_match_df = pd.read_csv('session_27_dataset/ipl_matches_2008_2020.csv', index_col = 'id')
ipl_match_df.head()
```

### 4. header paramter.

```python
# If in our dataset have messy issues not show clear column names then I can handle this by using header parameters
test_df = pd.read_csv('session_27_dataset/test.csv', header = 1)
test_df.head()
```

### 5. usecols parameter.

```python
# If I want to select just choses columns instead of working all dataset to store this in copy parameter then I can work on new dataset in which I selected columns
ipl_match_df = pd.read_csv('session_27_dataset/ipl_matches_2008_2020.csv')
ipl_match_df.head()
```

```python
# above dataset I selected chosen columns to work with 
ipl_match_df = pd.read_csv('session_27_dataset/ipl_matches_2008_2020.csv', usecols=['id', 'city', 'player_of_match', 'result'])
ipl_match_df.head()
```

### 6. skiprows parameter.

```python
# This parameter is used to skip those rows those are unnecessary in dataset while working on dataset.
ipl_match_df = pd.read_csv('session_27_dataset/ipl_matches_2008_2020.csv')
ipl_match_df.head()
```

```python
# We can see first 4 rows successfully skiped

ipl_match_df = pd.read_csv('session_27_dataset/ipl_matches_2008_2020.csv', skiprows= [0,1,2,3])
ipl_match_df.head()
```

```python
# I can also skiped rows by using of function in parameter. If we want to skiped even rows or odd or even manually rows I can do that
ipl_match_df = pd.read_csv('session_27_dataset/ipl_matches_2008_2020.csv', skiprows= lambda x : x in [1,2,3,4,5,6])
ipl_match_df.head()
```

### 7. nrows parameter.

```python
# Now If We want to work with just specific rows we can do that by using of nrows parameters.
ipl_match_df = pd.read_csv('session_27_dataset/ipl_matches_2008_2020.csv')
ipl_match_df.shape
```

```python
# here is just first 100 rows we selected from dataset 

ipl_match_df = pd.read_csv('session_27_dataset/ipl_matches_2008_2020.csv', nrows= 100)
ipl_match_df.shape
```

### 8. dtypes parameter. 

```python
# what if when in our dataset have different types of columns. We can also handle them to change their dtype so that our ml model can easily learn pattern 
pd.read_csv('session_27_dataset/ipl_matches_2008_2020.csv').info()
```

### 9. Handling Datetime categorical columns.

### 10. converter parameter. 

```python
def rename(name):
    if name == 'Punjab Cricket Association Stadium, Mohali':
        return 'PCASM'
    else: 
        return name
```

```python
rename('Punjab Cricket Association Stadium, Mohali') 
```

```python
# If We want to change long name into short abreviation word we can also do
ipl_match_df = pd.read_csv('session_27_dataset/ipl_matches_2008_2020.csv', converters={'venue': rename})
ipl_match_df.head()
```
