### Kernal Density Estimate(KDE)  :

#### KDE Kya Hai?
  KDE ek smooth curve hai jo data ka distribution dikhata hai.
  Yeh histogram ki tarah discrete nahi hota (matlab bins ya bars nahi hote), balki ek continuous smooth line hoti hai.
 Iska idea hai ki data kahan concentrated hai aur kahan spread hai.

### KDE Kyu Use Karte Hain?
Smooth Visualization:             - 
Histogram mein bars hote hain jo thoda rough dikhta hai, lekin KDE ek smooth curve deta hai jo distribution ko better dikhata hai

Continuous Data Ke Liye Be              - st:
Agar aapka data continuous hai (jaise height, weight, temperature), toh KDE uska distribution smooth tarike se dikhata ai.

Multiple Distributions Compare Karne Ke            -  Liye:
Agar aapko do ya zyada distributions ko ek saath compare karna hai, toh KDE ka use karo. Yeh overlap aur differences ko clear dikhata hai.

```python
# KDE plot :  Rather than using discrete bins, a KDE plot smooths the observations with a Gaussian kernel, producing a continuous density estimate
# by using of axes level function: Kde plot show in figure the probability that which percent this data have 

sns.kdeplot(data = tips, x = 'total_bill')
plt.show()
```

```python
# by using of figure level function

sns.displot( data = tips, kind = 'kde', x = 'total_bill', hue = 'sex', rug = True)
plt.show()
```

```python
sns.displot(data = tips , kind = 'kde', x = 'total_bill', hue = 'sex', height = 10, aspect = 6, fill = True, rug = True)
plt.show()
```

```python
# hue, fill, and aspect, height parameters

sns.displot(data = tips, kind='kde', x = 'total_bill', hue = 'sex', fill = True,height = 10, aspect = 3, rug = True)
plt.show()
```

```python
"""
Heatmap: Ek color-coded matrix hai jo data ko visualize karta hai.
Best Use: Correlation matrix, confusion matrix, ya kisi bhi tabular data ko dikhane ke liye.
"""

# some chatgpt examples :
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Example data
data = pd.DataFrame({
    'A': [1, 2, 3, 4],
    'B': [4, 3, 2, 1],
    'C': [1, 3, 2, 4]
})

# Correlation matrix
corr = data.corr()

# Heatmap
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Matrix Heatmap")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix

# Example data
y_true = [0, 1, 1, 0, 1, 0]
y_pred = [0, 1, 0, 0, 1, 1]

# Confusion matrix
cm = confusion_matrix(y_true, y_pred)

# Heatmap
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.title("Confusion Matrix Heatmap")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```
