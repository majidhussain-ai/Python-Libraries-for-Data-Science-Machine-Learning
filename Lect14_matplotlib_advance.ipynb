```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import gaussian_kde

# Sample data (e.g., website visitors ki daily count)
data = np.random.normal(1000, 200, 500)  # Mean = 1000, Std Dev = 200

# Histogram plot
plt.hist(data, bins=20, color='teal', edgecolor='black', density=True)

# Density plot overlay
density = gaussian_kde(data)
x_values = np.linspace(min(data), max(data), 1000)
plt.plot(x_values, density(x_values), color='red', label='Density Curve')

# Labels aur title
plt.xlabel('Visitors')
plt.ylabel('Density')
plt.title('Histogram with Density Plot')

# Show legend
plt.legend()

# Show plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.datasets import load_iris

# Iris dataset load karein
iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['species'] = iris.target

# Species ko names mein convert karein
df['species'] = df['species'].map({0: 'Setosa', 1: 'Versicolor', 2: 'Virginica'})

# Figure aur subplots banayein
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(12, 8))

# Plot 1: Sepal Length vs Sepal Width
for species in df['species'].unique():
    species_data = df[df['species'] == species]
    ax[0, 0].scatter(species_data['sepal length (cm)'], species_data['sepal width (cm)'], label=species)
ax[0, 0].set_title('Sepal Length vs Sepal Width')
ax[0, 0].set_xlabel('Sepal Length (cm)')
ax[0, 0].set_ylabel('Sepal Width (cm)')
ax[0, 0].legend()

# Plot 2: Petal Length vs Petal Width
for species in df['species'].unique():
    species_data = df[df['species'] == species]
    ax[0, 1].scatter(species_data['petal length (cm)'], species_data['petal width (cm)'], label=species)
ax[0, 1].set_title('Petal Length vs Petal Width')
ax[0, 1].set_xlabel('Petal Length (cm)')
ax[0, 1].set_ylabel('Petal Width (cm)')
ax[0, 1].legend()

# Plot 3: Sepal Length Distribution (Histogram)
for species in df['species'].unique():
    species_data = df[df['species'] == species]
    ax[1, 0].hist(species_data['sepal length (cm)'], alpha=0.5, label=species)
ax[1, 0].set_title('Sepal Length Distribution')
ax[1, 0].set_xlabel('Sepal Length (cm)')
ax[1, 0].set_ylabel('Frequency')
ax[1, 0].legend()

# Plot 4: Petal Length Distribution (Histogram)
for species in df['species'].unique():
    species_data = df[df['species'] == species]
    ax[1, 1].hist(species_data['petal length (cm)'], alpha=0.5, label=species)
ax[1, 1].set_title('Petal Length Distribution')
ax[1, 1].set_xlabel('Petal Length (cm)')
ax[1, 1].set_ylabel('Frequency')
ax[1, 1].legend()

# Layout adjust karein
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Step 1: X aur Y ke 1D arrays banayein
x = np.linspace(-5, 5, 100)  # X values (-5 se 5 tak)
y = np.linspace(-5, 5, 100)  # Y values (-5 se 5 tak)

# Step 2: meshgrid banayein
X, Y = np.meshgrid(x, y)

# Step 3: Z values calculate karein (e.g., z = sin(sqrt(x^2 + y^2)))
Z = np.abs(np.sin(np.sqrt(X**2 + Y**2)))

# Step 4: 3D plot banayein
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
p = ax.plot_surface(X, Y, Z, cmap='viridis')
plt.colorbar(p)

# Labels aur title
ax.set_xlabel('X Axis')
ax.set_ylabel('Y Axis')
ax.set_zlabel('Z Axis')
ax.set_title('3D Surface Plot using meshgrid')

# Show plot
plt.tight_layout()
plt.show()
```

```python
# Example 1: Basic 3D Surface Plot (Mathematical Function)

import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Step 1: X aur Y ke grids banayein
x = np.linspace(-5, 5, 100)  # X values (-5 se 5 tak)
y = np.linspace(-5, 5, 100)  # Y values (-5 se 5 tak)
X, Y = np.meshgrid(x, y)     # X aur Y ka grid banaya

# Step 2: Z values calculate karein (e.g., z = sin(sqrt(x^2 + y^2)))
Z = np.sin(np.sqrt(X**2 + Y**2))

# Step 3: 3D plot banayein
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')  # 3D plot initialize kiya
ax.plot_surface(X, Y, Z, cmap='viridis')    # Surface plot banaya

# Labels aur title
ax.set_xlabel('X Axis')
ax.set_ylabel('Y Axis')
ax.set_zlabel('Z Axis')
ax.set_title('3D Surface Plot: z = sin(sqrt(x^2 + y^2))')

# Show plot
plt.show()
```

```python
# heat distribution ka 3d plots
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Step 1: X aur Y ke grids banayein
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)

# Step 2: Z values calculate karein (e.g., heat distribution)
Z = np.exp(-(X**2 + Y**2))  # Gaussian function

# Step 3: 3D plot banayein
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='plasma')  # Plasma colormap use kiya

# Labels aur title
ax.set_xlabel('X Axis')
ax.set_ylabel('Y Axis')
ax.set_zlabel('Z Axis')
ax.set_title('3D Heat Distribution Plot')

# Show plot
plt.show()
```

### Contour Plots :

```python
"""
           Contour Plot Kya Karta Hai?
3D Data ko 2D Mein Visualize Karta Hai:
                                Contour plot 3D surface ko 2D plane mein represent karta hai, jisse patterns aur trends samajhna asaan ho jata hai.
Levels (Contour Lines) ko Dikhta Hai:
                                Har contour line ek specific Z value ko represent karti hai. Example: Agar Z = 10 hai, toh uss value wali contour line
                                dikhayi jayegi.
"""

# I can convert of surface first program into this shape basically contour plot convert of 3d plots into 2d contour lines(curves lines)
x = np.linspace(-10, 10, 100)
y = np.linspace(-10, 10, 100)
x,y =  np.meshgrid(x, y)

z = x**2 + y**2

fig = plt.figure()
ax = plt.subplot( )

ax.contour(x, y ,z, cmap = 'viridis')

plt.show()
```

```python
# I can also fill above graph by using of contourf function 
"""
Ess mein bhi aise hai jo kaam values wale hai wo konwe mein hai jiska color black nazar a rha hai aur jaise jaise values padhty gai hai 
to enn ka color change hota giya aur ye oppar ki taraf mail hoty gai

"""
x = np.linspace(-10, 10, 100)
y = np.linspace(-10, 10, 100)
x,y =  np.meshgrid(x, y)

z = x**2 + y**2

fig, ax = plt.subplots(figsize = (8,6))


p = ax.contourf(x, y ,z, cmap = 'viridis')
plt.colorbar(p)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Step 1: X aur Y ke grids banayein
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)

# Step 2: Z values calculate karein (e.g., z = x^2 + y^2)
Z = X**2 + Y**2

# Step 3: Contour plot banayein
plt.contour(X, Y, Z, levels=20, cmap='viridis')
plt.colorbar(label='Z Value')  # Colorbar add karein
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Basic Contour Plot: z = x^2 + y^2')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Step 1: X aur Y ke grids banayein
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)

# Step 2: Z values calculate karein (e.g., z = sin(x) + cos(y))
Z = np.sin(X) + np.cos(Y)

# Step 3: Filled contour plot banayein
plt.contourf(X, Y, Z, levels=20, cmap='plasma')
plt.colorbar(label='Z Value')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Filled Contour Plot: z = sin(x) + cos(y)')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Step 1: X aur Y ke grids banayein
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)

# Step 2: Z values calculate karein (e.g., z = x^2 - y^2)
Z = X**2 - Y**2

# Step 3: Contour plot banayein
contour = plt.contour(X, Y, Z, levels=10, cmap='coolwarm')
plt.clabel(contour, inline=True, fontsize=8)  # Contour lines par labels add karein
plt.colorbar(label='Z Value')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Contour Plot with Labels: z = x^2 - y^2')
plt.show()
```
