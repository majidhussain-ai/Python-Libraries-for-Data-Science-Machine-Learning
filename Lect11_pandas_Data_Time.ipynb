### Python Datetime() build_in function:

```python
# why separate objects to handle data and time when python already has datetime functionality?
```

#### Answer :
- syntax wise datetime is very convenient
- But the performance takes a hit while working with huge data. List vs Numpy Array
- The weaknesses of Python's datetime format inspired the NumPy team to add a set of native time series data type to NumPy.
- The datetime64 dtype encodes dates as 64-bit integers, and thus allows arrays of dates to be represented very compactly.

```python

date = np.array('2025-03-02', dtype=np.datetime64)
date
```

```python
date + np.arange(12)
```

#### important case :

- Because of the uniform type in NumPy datetime64 arrays, this type of operation can be accomplished much more quickly than if we were working directly with Python's datetime objects, especially as arrays get large 

- Pandas Timestamp object combines the ease-of-use of python datetime with the efficient storage and vectorized interface of numpy.datetime64

- From a group of these Timestamp objects, Pandas can construct a DatetimeIndex that can be used to index data in a Series or DataFrame

### Date_range() function :

```python
d1 = pd.to_datetime("2025-09-01 14:00:00")
d2 = pd.to_datetime("2025-09-03 16:30:00")

diff = d2 - d1
print(diff)
print(diff.days)      # pure days
print(diff.seconds)   # extra seconds
print(diff.total_seconds())  # total duration in seconds
```

### to_datetime() function :

```python
"""
Define:
      The to_datetime() function in Pandas is used to convert a variety of date formats into a standard datetime64 format.
      It is highly useful in time-series analysis, data preprocessing, and handling messy date formats.

Syntax :
        pd.to_datetime(arg, format=None, errors='raise', utc=False, dayfirst=False, yearfirst=False, exact=True, unit=None, infer_datetime_format=False)

"""

# Convert a series string data format into datetime
df = pd.Series(['2025/2/3', '2023/4/9', '2025/5/8'])
converts_dates = pd.to_datetime(df)
converts_dates
```

```python
# converts into different formats 

dates = pd.Series(["2024-03-01", "2024/3/2", "March 3, 2024"])
converted_dates = pd.to_datetime(dates, yearfirst = True, errors = 'coerce')
print(converted_dates)
```
