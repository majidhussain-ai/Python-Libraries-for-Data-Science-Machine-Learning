### some functions like concate, merge :

```python
"""
concat ka matlab hai rows ya columns ko simply jodna (stack karna).
Isay tum tab use karte ho jab dataframes ka structure same ho (same columns) aur bas tumhe unko jorna hai.

Syntax:
pd.concat([df1, df2], axis=0)


axis=0 â†’ rows ko jor dega (default)
axis=1 â†’ columns ko jor dega
"""
```

```python
"""
2. pd.merge() â†’ Database-style Join

Ye SQL JOIN ki tarah kaam karta hai (INNER, LEFT, RIGHT, OUTER).
Tum matching keys/columns ke base par DataFrames combine karte ho.
syntax:
        pd.merge(df1, df2, on="common_column", how="inner")

"""
```

```python
"""
ðŸ”¹ 1. join() ka basic idea

join() ka use 2 ya usse zyada DataFrames ko join karne ke liye hota hai.
Ye index ke basis par DataFrames ko join karta hai (default behavior).
merge() aur concat() se different hai, kyunki join() mainly index alignment karta hai.
"""
```

```python
# anywhere we need to concate two or more dataset to perform operation on it so we can do it by uisng of concat() function
# aur ignore_index = True ess liye lagaya take index ki sequnce maintain rhe
result = pd.concat([nov, dec], ignore_index = True)
result
```

```python
# multi_index dataframe 
# vertically stacking
"""
ye ess liye kar rhe hai keh hamein every dataset ke khud ke index apne name ke sath show hon
aur hamein koi confusion na rhe 
"""

multi_index = pd.concat([nov, dec], keys = ['nov', 'dec'])
multi_index
```

```python
# from multi_index dataframe to fetch any one data we need. Here is one data of Nov we pass keys list name
multi_index.loc['nov']
```

```python
new_regs = pd.concat([nov, dec], keys = ['nov', 'dec']).reset_index()
new_regs
```

```python
# we can also do to fetch any specfic row: This is fetching data by label wise 
multi_index.loc[('nov',4)]
```

```python
# this fetching data by index wise but the result is same as above with loc label index
multi_index.iloc[4]
```

```python
# horizontaly stacking by using of axis we do rows wize but bydefault is columns wize
pd.concat([nov, dec], axis =1)
```

```python
df1 = pd.DataFrame({
    "emp_id": [1, 2, 3],
    "name": ["Ali", "Sara", "John"]
})

df2 = pd.DataFrame({
    "id": [1, 2, 3],
    "dept": ["IT", "HR", "Finance"]
})

result = pd.merge(df1, df2, left_on="emp_id", right_on="id", how="inner")
print(result)
```

```python
df1 = pd.DataFrame({
    "id": [1, 2, 3],
    "year": [2020, 2021, 2020],
    "name": ["Ali", "Sara", "John"]
})

df2 = pd.DataFrame({
    "id": [1, 2, 3],
    "year": [2020, 2021, 2020],
    "salary": [50000, 60000, 70000]
})

result = pd.merge(df1, df2, on=["id","year"], how="inner")
print(result)
```

```python
result = pd.merge(df1, df2, on=["id","year"], how="outer", indicator=True)
print(result)
```

```python
df1 = pd.DataFrame({"key": ["A", "B", "C"], "val1": [1, 2, 3]})
df2 = pd.DataFrame({"key": ["B", "C", "D"], "val2": [4, 5, 6]})

result = pd.merge(df1, df2, how="outer")
print(result)
```

```python
print(students.head(2).reset_index())
print('\n\nNew dataset\n',new_regs.head(2))
```

```python
# inner joining..
"""
Hmm yahan inner joining, outer, left, right joining ess liye use karte hain kio keh jbb hamare pass do columns hain like student_id aur course_id to en
mein student ka name nhy ho ga. then by using of joining we can do this keh on the basis of student_id we can merge student_id with student_name this
can do only on merging .
like below I have student dataset which is student_id and course_id i want to merge on the base of student_id , student_name i can do this very easy
"""
"""
Hamare pass two month ke course ka data set hai jo keh new_regs mein store. lkn hamare students 25 hain jinho ne 58 courses kharede 
ye dono elahda elahda dataset mein store hain aur mein chahta ho keh en dono mein sy . Every students ka name with his course ID ke name sy show ho
to mein ess ke liye data ko merge karna pare ga
"""


# Ye on dataset ko join kare ga jn mein common ids aur data ho ga
pd.merge(students,new_regs, on = 'student_id', how = 'inner')
```

```python
# left joining......
"""
Ye left joining ess liye use karte hai take hamein left ke all data mile irrespective of right 
ess sy hamein aisa dta milta hai jo complete left mein hota hai
aur eidhar course_id ke base pr on ess liye laga rhe hai keh ye dono mein common hai

new_regs ke column level_0 and level_1 ess liye empty hain kio keh en cources(c++ and numpy)mein ksi ne ab tk enrol nhy kiya 
"""

# Ess left joining mein bhi aise hee all left dataset columns and common on both 
total_reg = pd.merge(cources, new_regs, on = 'course_id', how = 'left')
total_reg
```

```python
# right join
temp_df = pd.DataFrame({
    'student_id':[26,27,28],
    'name':['Nitish','Ankit','Rahul'],
    'partner':[28,26,17]
})

students = pd.concat([students,temp_df],ignore_index=True)
```

```python
# this is maximum student_id index column
new_regs[new_regs['student_id'] == 50]
```

```python
# right joining.....
"""
Ess mein aisa data show ho ga jinho ne course mein enrol to kiya magar koi course nhy khrida
Es output ke last mein three aise student hai jinho ne enrol kiya magar koi course nhy karida
"""

# This is also all right dataset columns and common on both
pd.merge(students,new_regs, on = 'student_id', how= 'right')
```

```python
# outer joining....
"""
ess mein kuch aise students hai jinho ne course_id mein to enrol nhy kiya magar website pr enrol kiya jaise khud ko sirf register karte hai magar koi course 
nhy khridte magar kuch aise student bhi hain jinki course_id hai magar onn ke name mention nhi hain
"""
pd.merge(students, new_regs, on = 'student_id', how = 'outer').tail(10)
```
