### find ranking of each movie in the group according of 'imdb_rating'

```python
# Here group is basically a whole dataframe or dataset
# find ranking of each movie in the group according to IMDB score

def rank_movie(group):
  group['genre_rank'] = group['imdb_rating'].rank(ascending=False, pct = True)
  return group
```

```python
genres.apply(rank_movie)
```

### find normalized IMDB rating group wise:

```python
# basically here i'm also using the formula of normalization which is min_max normalization used for accuracey fast resutl and give values in 0,1
# find normalized IMDB rating group wise
def normal(group):
    group['normal_rating'] = (group['imdb_rating']-group['imdb_rating'].min())/(group['imdb_rating'].max() - group['imdb_rating'].min())
    return group
genres.apply(normal)
```

```python
# By using of this function i can find any batsman highest runs by pass of his name

# Create a function that can return the highest score of any batsman
def highest(batsman):
    temp = ipl[ipl['batsman'] == batsman]
    return temp.groupby('match_id')['batsman_runs'].sum().sort_values(ascending = False).head(1).reset_index()
```
