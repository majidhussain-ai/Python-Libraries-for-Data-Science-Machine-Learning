```python
# common functions used for str operations
# upper/lower/capitalize/title
df['Name'].str.upper()
df['Name'].str.lower()
df['Name'].str.capitalize()
df['Name'].str.title()
df['Name'].str.swapcase()
```

```python
# separate name and store it into new columns in dataset
df['Sir_name'] = df['Name'].str.split(',').str.get(0)
df.head()

df['new_name'] = df['Name'].str.split(',').str.get(1)
df.sample(4)
```

```python
# From above question i split names into three part like Sir_name, Title, and Own_name now i want to create new column of title_name and Own_name

df[['Title_name', 'Own_name']]= df['Name'].str.split(',').str.get(1).str.strip().str.split('.', n = 1, expand = True)
df.head(2)
```

```python
# Ess mein Miss , Ms, aur Mlle same words hain mein en ko ek karna chahta ho
df['Title_name'].value_counts()

df['Title_name'] = df['Title_name'].str.replace('Ms.' , 'Miss.')
df['Title_name'] = df['Title_name'].str.replace('Mlle.' , 'Miss.')
```

```python
df['Title_name'].value_counts()
```

```python
# Hamein name dataset column mein sy wo name find karne hain jo A sy start howe hain
df[df['Own_name'].str.startswith('A')].sample(2)
```

```python
# contain() : function to find or search you want from string
# I want to find john name from Own_name column of dataset. Here case = False means you find value in large words or capital on both

df[df['Own_name'].str.contains('John' , case = False)].head(2)
```

```python
# Find those names those starts and ends with vowel

df[df['Own_name'].str.contains('^[aeiouAEIOU].+[aeiouAEIOU]$')].head(3)
```

```python
# find those names who start and ends with consunant 
df[df['Own_name'].str.contains('^[^aeiouAEIOU].+[^aeiouAEIOU]$')].head(4)
```

```python
import pandas as pd

data = pd.Series(["machine learning", "Deep Learning", "artificial intelligence", None])

print(data.str.upper())     # Convert to uppercase
print(data.str.lower())     # Convert to lowercase
print(data.str.capitalize())  # Capitalize the first letter
print(data.str.title())     # Capitalize first letter of each word
print(data.str.swapcase())  # Swap upper and lower case
```

```python
print(data.str.contains("learning", case=False, na=False))  # Case insensitive check
print(data.str.startswith("Deep"))
print(data.str.endswith("intelligence"))
```

```python
reviews = pd.Series([
    "  This product is AMAZING!   ",
    "Worst experience ever...  ",
    "Absolutely fantastic!! ",
    None
])

cleaned_reviews = reviews.str.strip().str.lower().str.replace(r'[^\w\s]', '', regex=True)
print(cleaned_reviews)
```
