```python
# Load a dataset
data = {
    'Age': [25, 30, np.nan, 40],
    'Salary': [50000, np.nan, 70000, 80000],
    'Department': ['HR', 'IT', 'IT', np.nan]
}
df = pd.DataFrame(data)
# Fill missing values
df['Age'].fillna(df['Age'].mean(), inplace=True)  # Fill Age with mean
df['Salary'].fillna(0, inplace=True)  # Fill Salary with 0
df['Department'].fillna('Unknown', inplace=True)  # Fill Department with 'Unknown'

print("Processed DataFrame:\n", df)
```

### apply(s + d):

```python
"""
define:
            The apply() function is used to apply a function along an axis (rows or columns) of a DataFrame or to each element of a Series.
              It’s a versatile tool for transforming data

use cases:
          3. Use Case in Machine Learning
         In machine learning, apply() is commonly used for:

          Feature engineering: Create new features based on existing ones.
          Data preprocessing: Transform or clean data (e.g., scaling, encoding).
         Custom calculations: Perform row-wise or column-wise calculations.
"""
# Load a dataset
data = {
    'Age': [25, 30, 35, 40],
    'Salary': [50000, 60000, 70000, 80000]
}
df = pd.DataFrame(data)

# Create a new feature 'Salary_Category'
def categorize_salary(salary):
    if salary < 60000:
        return 'Low'
    elif salary < 80000:
        return 'Medium'
    else:
        return 'High'

df['Salary_Category'] = df['Salary'].apply(categorize_salary)
print("DataFrame with 'Salary_Category':\n", df)
```
