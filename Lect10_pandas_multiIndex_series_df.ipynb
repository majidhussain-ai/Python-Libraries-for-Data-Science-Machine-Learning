### MultiIndex with syntax and basics example:

```python
"""
ðŸ”¹ 1. MultiIndex kya hai?

ðŸ‘‰ Normally DataFrame ya Series ka ek single index hota hai (jaise 0,1,2... ya students ke names).
ðŸ‘‰ Lekin kabhi humein 2 ya 3 levels ke index chahiye hote hain â€” tab hum MultiIndex use karte hain.

Use cases:
            Students ke subjects aur test type (Mid/Final) ek saath index mein rakhna.
            Companies ke Year aur Quarter ek saath.
            Machine Learning mein feature groups (cat, num, text) ko multi-level index dena.
Syntax:
            pandas.MultiIndex.from_arrays(arrays, names=None)
            pandas.MultiIndex.from_tuples(tuples, names=None)
            pandas.MultiIndex.from_product(iterables, names=None)

"""

# With the help of 1'st syntax
subjects = ['Math','Math','Machine learning', 'Machine learning']
semester = ['Spring', 'Fall', 'Spring', 'Fall']
multi_index = pd.MultiIndex.from_arrays([subjects, semester], names = ['Subject', 'Semester'])
grades = pd.DataFrame({'marks':[89,98,23,21]}, index = multi_index)
grades
```

```python
# getting values from dataset
grades.loc['Math']['Fall']
```

```python
# Now With the help of 2nd Syntax
# Tuples banaye (Country, City)
index = pd.MultiIndex.from_tuples(
    [('USA', 'New York'), ('USA', 'Chicago'), ('Canada', 'Toronto'), ('Canada', 'Vancouver')],
    names=['Country', 'City']
)

# DataFrame create
population = pd.DataFrame({'Population (M)': [8.3, 2.7, 2.9, 2.6]}, index=index)
population
```

```python
population.loc['Canada']
```

```python
# With the help of 3rd Syntax
# Product of Items Ã— Sizes
index = pd.MultiIndex.from_product(
    [['Shirt', 'Pants'], ['Small', 'Medium', 'Large']],
    names=['Item', 'Size']
)

# Inventory Data
inventory = pd.DataFrame({'size number':[50, 30, 20, 40, 25, 15]}, index=index)
inventory
```

```python
inventory.loc['Shirt']['Small']
```

### long vs wide data :

#### 1. Wide Format : 
- Is format mein har variable ke liye alag column hota hai.
- Rows zyada â€œspreadâ€ hoti hain.
- Easy to read for humans, lekin analysis/ML ke liye kabhi kabhi tough hota hai.

#### 2. Long Format
- Is format mein ek hi variable ke multiple values rows ke andar hoti hain.
- Zyada rows hoti hain, lekin columns kam.
- ML/DL mein zyadatar long/tidy format prefer hota hai.

```python
# Wide Format with syntax and parameter explain
"""
Wide --> Long using pd.melt()
Syntax:
        pd.melt(frame, id_vars=None, value_vars=None, 
        var_name=None, value_name="value", col_level=None)

Parameters Explanation:

                    frame â†’ DataFrame jo aapke paas hai (wide format).
                    id_vars â†’ woh column(s) jo identifier banenge (fixed rahenge).
                    value_vars â†’ woh columns jo "unpivot" karke ek hi column mein le aayenge. Agar None ho toh saare non-id_vars le lega.
                    var_name â†’ naye column ka naam jisme melted variable store hoga.
                    value_name â†’ naye column ka naam jisme values store hongi.
                    col_level â†’ agar multi-level columns hain (MultiIndex), toh specify karte ho kis level ko melt karna hai.

"""
```

```python
# Long format to Wide format using pivot and pivot_table.
"""
Sntax:
       DataFrame.pivot(index=None, columns=None, values=None)
Parameters Explanation:
                        index â†’ woh column jo rows banega.
                        columns â†’ woh column jo naye column headers banayega.
                        values â†’ woh column jo cell ki values provide karega.
                        âš¡ Note: Agar duplicate combinations ho toh pivot_table() use karo aur aggfunc specify karo.

"""
```

```python
# melt() : Ess ka kaam hai ke ye wide data format ko long data format mein change kar dita hai
# melt -> simple example branch
# wide to long
pd.DataFrame({'cse':[120]}).melt()
```

```python
# melt -> branch with year
pd.DataFrame({'cse':[120],'ece':[100],'mech':[50]}).melt(var_name='Branch',value_name='Students')
```

### pivot table :

```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
# ['tips'] dataset : bydefault ek seaborn ka dataset hai jo phele sy ess mein import shuda hai
df = sns.load_dataset('tips')
df.head()
```

```python
"""
From above data i apply groupby function to separate bill separate for male and separate for female and then i apply on it that which expensive 
for male and for female and then i use sum() function on it to calculate all expensive
"""
df1 = df.groupby('sex', observed = True)['total_bill'].sum()
df1
```

```python
# Here I'm using of unstack() which means if our dataset output in long format and not easily to read so we can convert it in wide format with use of unstack()

df.groupby(['sex', 'smoker','day'],  observed = True)[['total_bill']].mean().unstack()
```

```python
# this is like of above data but here is using of pivot_table which is more flexible than groupby by using of groupby there are different ways to where 
# to use it for better performance. pivot_table of agg_func bydefualt is mean if you want to sum of these valuse then you can do like below
# pivot_table ess liye use karte hain keh agar hamare dataset mein values same hain :

df.pivot_table(index = ['sex', 'smoker'], columns = 'day', values = 'total_bill', observed = True)
```

```python
# There is the flexibility by using of pivot_table of aggfunc

df.pivot_table(index=['sex','smoker'], columns = 'day', values = 'total_bill', observed=False, aggfunc='std')
```

```python
"""
Here I apply groupby function on two columns sex mean male and female and smoker which means Yes are smokers and No mean non_smoker 
"""
df.groupby(['sex','smoker'],  observed=False)[['total_bill']].mean().unstack()
```

```python
"""
Like above groupby function I can also apply pivot_table also give result above but this pivot_table is more advance for flexibility and mostly used 

pivot_table Syntax :
                     f.pivot_table(index='column1', columns='column2', values='column3', aggfunc='function')
Parameter	Description
           index:	Jo column rows mein dikhana hai
         columns:	Jo column separate categories ke liye use hoga
          values:	Jis column ke values aggregate karni hain
        aggfunc :	Kaunsi function lagani hai (sum, mean, count, etc.)
                   
"""
df.pivot_table(index='sex',columns='smoker',values='total_bill', observed = True)
```

```python
# aggfunc and margins 
df.pivot_table(index='sex',columns='smoker',values='total_bill',observed = True,aggfunc='sum', margins = True)
```

```python
# by using of margins if we use aggfunc = 'mean' then the margins all values will also be in mean if we use aggfunc = 'sum', then margins values also sum

(1337.07 + 1919.75)
```

```python
# all cols together
df.pivot_table(index='sex',columns='smoker', values = 'total_bill', aggfunc = 'mean', observed = True, margins = True)
```

```python
# multidimensional
"""
Here i used pivot_table function and then perfom on it multidimentional aggfunc to separatly apply on columns and rows: and last margins function to count
all values sum rows wise and columns wise
"""
df.pivot_table(index=['sex','smoker'],columns=['day','time'],aggfunc={'size':'mean','tip':'max','total_bill':'sum'}, observed=True, fill_value = 1)
```

```python
# margins
df.pivot_table(index='sex',columns='smoker',values='total_bill',aggfunc='sum',margins=True, observed = True)
```

```python
gdp_wide = pd.DataFrame({
    "Country": ["Pakistan", "India", "China"],
    "2022": [350, 2800, 16000],
    "2023": [370, 2900, 16500],
    "2024": [400, 3100, 17000]
})

# Wide â†’ Long
gdp_long = pd.melt(gdp_wide, id_vars="Country", 
                   var_name="Year", value_name="GDP")

print("\nLong Format:\n", gdp_long)

# Long â†’ Wide
gdp_back = gdp_long.pivot(index="Country", 
                          columns="Year", 
                          values="GDP").reset_index()

print("\nWide Format Back:\n", gdp_back)
```

```python
import pandas as pd  

data = {
    'City': ['Mumbai', 'Delhi', 'Mumbai', 'Delhi'],  
    'Product': ['Phone', 'Laptop', 'Laptop', 'Phone'],  
    'Sales': [100, 200, 150, 50]  
}  

df = pd.DataFrame(data)  

# Pivot Table (Sum of Sales by City and Product)  
pivot = pd.pivot_table(
    df,  
    values='Sales',  
    index='City',  
    columns='Product',  
    aggfunc='sum',  
    fill_value=0  
)  

pivot.plot(kind = 'bar')
```

```python
df['Year'] = [2023, 2023, 2024, 2024]  

# Multi-level Pivot (Year + Product vs City)  
pivot_multi = pd.pivot_table(
    df,  
    values='Sales',  
    index='City',  
    columns=['Year', 'Product'],  
    aggfunc='sum',  
    fill_value=0  
)  
df
```

```python
# Adding a 'Year' column  
df['Year'] = [2023, 2023, 2024, 2024]  

# Multi-level Pivot (Year + Product vs City)  
pivot_multi = pd.pivot_table(
    df,  
    values='Sales',  
    index='City',  
    columns=['Year', 'Product'],  
    aggfunc='sum',  
    fill_value=0  
)  

pivot_multi.plot(kind = 'bar')
```
