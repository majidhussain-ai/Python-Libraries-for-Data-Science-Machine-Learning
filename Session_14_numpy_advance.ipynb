### Broadcasting :


- The term broadcasting describes how numpy arrays treats with different shape of arrays during arithematic operations.
- The smallar array is " broadcast " across the larger array so that they have compatible shapes.

```python
# same shape
import numpy as np
a1 = np.arange(6).reshape(2,3)
a2 = np.arange(6,12).reshape(2,3)
print(a1)
print(a2)
print(a1 + a2)

# Adding would be item wize
```

```python
# diff shape 
a1 = np.arange(6).reshape(2,3)
a2 = np.arange(3)
print(a1)
print(a2)
print(a1 + a2)
```

### Broadcasting Rules :
#### 1. Make two arrays have the same number of diminsion.
- If the numbers of diminsions of two arrays are different, add new diminsion of size 1 to the head of array with the smaller diminsion.

#### 2. Make each diminsion of the two arrays with same size.
- If the size of each diminsion of two arrays do not match,  diminsion with size 1 are stretched to the size of other array.
- If there is a diminsion whose size is not 1 , it cannot be broadcast and an error raised.

### Examples of broadcasting :

```python
# it is [ (4 * 3) & (3,)] diminsion
a1 = np.arange(12).reshape(4,3)
a2 = np.arange(3)
print(a1)
print(a2)
print(a1 + a2)
```

### Binary Cross entropy :

```python
"""
Binary Cross-Entropy (also known as log loss) is a loss function commonly used in binary classification problems.
It measures the performance of a classification model whose output is a probability value between 0 and 1.
The formula for Binary Cross-Entropy is:
Averege loss for multiple samples
bse = -1/N âˆ‘ (yi.log(^yi) + (1 -yi).log(1-^yi))
"""
import numpy as np

def binary_cross_entropy(y_true, y_pred, epsilon=1e-15):
    """
    Calculate the Binary Cross-Entropy (log loss) between true labels and predicted probabilities.

    Parameters:
    y_true (array-like): Actual binary labels (0 or 1).
    y_pred (array-like): Predicted probabilities (between 0 and 1).
    epsilon (float): Small value to avoid log(0).

    Returns:
    float: The Binary Cross-Entropy loss.

    numpy.clip(a, a_min, a_max, out=None)
    """
    # Ensure inputs are numpy arrays
    y_true = np.array(y_true)
    y_pred = np.array(y_pred)
    
    # Clip predictions to avoid log(0) or log(1)
    y_pred = np.clip(y_pred, epsilon, 1 - epsilon)
    
    # Calculate the Binary Cross-Entropy
    bce = -np.mean(y_true * np.log(y_pred) + (1 - y_true) * np.log(1 - y_pred))
    
    return bce

# Example usage
y_true = [1, 0, 1, 1, 0]
y_pred = [0.9, 0.1, 0.8, 0.7, 0.2]

loss = binary_cross_entropy(y_true, y_pred)
print(f"Binary Cross-Entropy Loss: {loss:.3f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Sigmoid Function Define Karna
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# X Values Generate Karna
x = np.linspace(-10, 10, 100)
y = sigmoid(x)

# Plot Sigmoid Function
plt.plot(x, y, label="Sigmoid Function")
plt.xlabel("Input (x)")
plt.ylabel("Output (Ïƒ(x))")
plt.title("Sigmoid Function Curve")
plt.legend()
plt.grid()
plt.show()
```

```python
def sigmoid_derivative(x):
    return sigmoid(x) * (1 - sigmoid(x))

# Derivative Plot Karna
x = np.linspace(-10, 10, 100)
y = sigmoid_derivative(x)

plt.plot(x, y, label="Sigmoid Derivative", color='red')
plt.xlabel("Input (x)")
plt.ylabel("Derivative (Ïƒ'(x))")
plt.title("Sigmoid Derivative Curve")
plt.legend()
plt.grid()
plt.show()
```

```python
def predict(x, weights):
    z = np.dot(x, weights)
    return sigmoid(z)

# Example Data
x = np.array([1, 2])  # Features
weights = np.array([0.5, -0.5])  # Weights

# Prediction
output = predict(x, weights)
print(f"Prediction Probability: {output:.4f}")
```

```python
import numpy as np

def mean_squared_error(y_true, y_pred):
    errors = y_true - y_pred  # Difference between actual & predicted
    squared_errors = errors ** 2  # Square errors
    mse = np.mean(squared_errors)  # Take the mean
    return mse

# Example Data
y_actual = np.array([3, -0.5, 2, 7])
y_predicted = np.array([2.5, 0.0, 2, 8])

# Calculate MSE
mse_value = mean_squared_error(y_actual, y_predicted)
print(f"Mean Squared Error: {mse_value:.4f}")
```

```python
import matplotlib.pyplot as plt

# Errors Calculate Karna
errors = y_actual - y_predicted
squared_errors = errors ** 2

# Plot
plt.figure(figsize=(8,5))
plt.bar(range(len(errors)), squared_errors, color='blue', alpha=0.6, label="Squared Errors")
plt.axhline(y=mse_value, color='r', linestyle='--', label="MSE")

plt.xlabel("Data Points")
plt.ylabel("Squared Error")
plt.title("Squared Errors and MSE")
plt.legend()
plt.grid()
plt.show()
```
