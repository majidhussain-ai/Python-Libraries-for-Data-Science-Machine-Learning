#### EDA:

Exploratory Data Analysis (EDA) is a crucial step in the data analysis process that involves summarizing the main characteristics of a dataset, often using visual methods. It is an approach to analyzing datasets to summarize their main characteristics, often with visualizations, before applying more formal modeling or hypothesis testing.

#### Purpose of EDA:

1. **Understand the Dataset:**
      - Gain insights into the structure, patterns, and relationships within the data.
      - Identify the distribution of variables, missing values, and outliers.
2. **Data Cleaning:**

2. **Generate Hypothises**:

### Scatter Plot() :

```python
"""
Scatter plot ek 2D visualization technique hai jo do numerical variables ke beech relationship ko show karta hai. 
Har point ek (X, Y) coordinate ko represent karta hai.

Scatter plots numerical data points ko individual dots ke form mein dikhate hain. Ye data ke patterns, clusters, 
aur relationships (jaise correlation) ko samajhne ke liye use hote hain.

Use Case:
Numerical Data: Relationships between two variables.

Syntax :
         import matplotlib.pyplot as plt
         plt.scatter(x, y, s=None, c=None, marker=None, cmap=None, alpha=None, linewidths=None, edgecolors=None)
         plt.show() # to display the plot
"""
x = np.linspace(-10, 10, 50)
size = np.random.randint(1,100, 50)
y = 10*x+4+np.random.randint(1, 300, 50)
y
```

```python
# simple scatter plot creating. aplha parameter is used for transperancy which means light color of your scatter points
plt.scatter(x, y, s = size, marker = 's', alpha = 0.8)
```

```python
""" 
Here and every scatter plot where we use cmap and colorbar plot. The same points represent the same index like 2,2,2 index but with different values 
the value of these same index might be like 99, 88, 78 . In the given below graph same index is 7 hours three time in the dataset but with different values
first is (7i, to 86v), (7i to 88v), (7i to 94v) here i mean index and v means values
"""

x = [5,7,8,7,6,9,5,6,7,8]
y = [99,86,87,88,100,86,103,87,94,78]

plt.scatter(x, y, s = y, c = x, cmap = 'viridis')
plt.xlabel("Study Hours")
plt.ylabel("Marks")
plt.title("Study Hours vs Marks (Correlation Check)")
plt.colorbar()
plt.show()
```

```python
# Stacked bar graph

import matplotlib.pyplot as plt  

# Data for the bar graph  
days = ['Thur', 'Fri', 'Sat', 'Sun']  
male_tips = [89, 27, 182, 187]  
female_tips = [82, 25, 78, 61]  

# Plotting the bar graph  
fig, ax = plt.subplots(figsize=(8, 6))  

# Create bars for male tips  
ax.bar(days, male_tips, label='Male', color='#2cb7d1')  

# Create bars for female tips, stacked on top of the male tips  
ax.bar(days, female_tips, bottom=male_tips, label='Female', color='#9c27b0')  

# Adding labels and title  
ax.set_title('Tips by Day and Gender', fontsize=14)  
ax.legend()  

# Adding value labels on top of the bars  
for i, (m, f) in enumerate(zip(male_tips, female_tips)):  
    total = m + f  
    ax.text(i, total + 3, str(total), ha='center', va='bottom', fontsize=12)  
    ax.text(i, m + f/2, str(f), ha='center', va='center', color='white', fontsize=12)  
    ax.text(i, m/2, str(m), ha='center', va='center', color='white', fontsize=12)  

# Setting background color to black  
fig.patch.set_facecolor('black')  
ax.set_facecolor('black')  

# Setting text color to white  
ax.title.set_color('white')  
ax.xaxis.label.set_color('white')  
ax.yaxis.label.set_color('white')  
ax.tick_params(axis='x', colors='white')  
ax.tick_params(axis='y', colors='white')  
ax.legend(labelcolor='white')  

# Adding gridlines  
ax.grid(axis='y', linestyle='-', alpha=0.2, color='white')  

# Removing spines  
ax.spines['top'].set_visible(False)  
ax.spines['right'].set_visible(False)  
ax.spines['left'].set_visible(False)  
ax.spines['bottom'].set_color('white')  

# Display the graph  
plt.tight_layout()  
plt.show()  
```

```python
import matplotlib.pyplot as plt  
import pandas as pd  
import seaborn as sns  #Import seaborn for sample data  

# Load the tips dataset (or replace with your data loading)  
tips = sns.load_dataset('tips')  

fig, axes = plt.subplots(1, 4, figsize=(18, 5)) #Increased figure size for better readability.  

# Total Bill  
axes[0].hist(tips['total_bill'], bins=20, color='skyblue', edgecolor='black')  
axes[0].set_title('Total Bill Distribution')  
axes[0].set_xlabel('Total Bill ($)')  
axes[0].set_ylabel('Frequency')  

# Tip  
axes[1].hist(tips['tip'], bins=15, color='orange', edgecolor='black')  
axes[1].set_title('Tip Distribution')  
axes[1].set_xlabel('Tip ($)')  
axes[1].set_ylabel('Frequency')  

# Size  
axes[2].hist(tips['size'], bins=5, color='purple', edgecolor='black')  
axes[2].set_title('Size Distribution')  
axes[2].set_xlabel('Size (Number of People)')  
axes[2].set_ylabel('Frequency')  

# Day (Handling Categorical Data)  
day_counts = tips['day'].value_counts()  
axes[3].bar(day_counts.index, day_counts.values, color='pink', edgecolor='black')  
axes[3].set_title('Day Distribution')  
axes[3].set_xlabel('Day')  
axes[3].set_ylabel('Frequency')  

plt.tight_layout()  
plt.show()  
```
