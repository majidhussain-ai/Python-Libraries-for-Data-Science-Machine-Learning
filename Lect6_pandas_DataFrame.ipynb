### str.contains method:
- Ye Pandas ka string accessor method hai jo Series (DataFrame ke ek column) par apply hota hai.
- Iska kaam hai: check karna ke ek string ke andar ek chhoti substring ya pattern exist karti hai ya nahi.
#### Syntax: 
- df['column'].str.contains('pattern', case=True, na=False, regex=True)
#### Parameters:
- pattern → search karna wala text ya regex
- case → case-sensitive ya nahi (default True)
- na → NaN values ko kis tarah handle kare (default NaN, set False to ignore)
- regex → agar True ho to pattern ko regex samjha jata hai (default True)


```python
import pandas as pd

df = pd.DataFrame({
    'Names': ['Majid Hussain', 'Ali Khan', 'Sara Sheikh', 'MAJID Ali']
})

df['Names'].str.contains('Majid')
```

```python
# Here case = True means doesn't matter your pattern in upper case or lower case if it match then it count in True ('majid' == 'MAJID')
df['Names'].str.contains('majid', case = False)
```

```python
df[df['Names'].str.contains('Ali')]
```

```python
print(df['Names'].str.contains('^Majid'))  # names starting with "Majid"
print(df['Names'].str.contains('Ali$'))     # names ending with "Ali"
```
